**Workflow:** Edit locally → `git push` → run in Colab.  
Cell 1 clones (or updates) the repo from GitHub so you always get the latest code.

```python
# ── 1. Clone / update the repo from GitHub ────────────────────────────────
import os

REPO_URL = 'https://github.com/YosemiteSamurai/CurrentPrediction.git'
REPO_DIR = '/content/CurrentPrediction'

if os.path.isdir(os.path.join(REPO_DIR, '.git')):
    print("Repo already cloned — pulling latest changes...")
    os.system(f'git -C {REPO_DIR} pull')
else:
    print("Cloning repo...")
    os.system(f'git clone {REPO_URL} {REPO_DIR}')

print("Done.")
```

```python
# ── 2. Set up Python path ─────────────────────────────────────────────────
import sys, os

gcn_dir = '/content/CurrentPrediction/gcn'
assert os.path.isdir(gcn_dir), f"gcn/ not found at {gcn_dir} — did Cell 1 finish successfully?"

if gcn_dir not in sys.path:
    sys.path.insert(0, gcn_dir)

os.chdir(gcn_dir)
print(f"Working directory: {os.getcwd()}")
```
